```python
import requests
import json
from pprint import pprint
import pandas as pd
```

```python
# Toronto Police â€“ Serious/Fatal Traffic Incidents API
base_url = 'https://services.arcgis.com/S9th0jAJ7bqgIRjw/arcgis/rest/services/KSI/FeatureServer/0/query?where=1%3D1&outFields=*&outSR=4326&f=json'
```

```python
data = requests.get(base_url).json()
pprint(data['features'][3])
```

```python
pprint(data['features'][3]['attributes']['ACCLASS'])
pprint(len(data['features']))
```

```python
# state the lists
acc_class = []
acc_loc = []
acc_district = []
acc_date = []
acc_division = []
acc_division_code = []
acc_invtype = []
acc_impactype = []
acc_injur = []
acc_invage = []
acc_latitute = []
acc_longitue = []
acc_light = []
acc_road_condition = []
acc_traffic = []
acc_time = []
acc_visibility = []
acc_year = []
acc_hood_id = []
```

```python
for i in range(len(data['features'])):
    acc_class.append(data['features'][i]['attributes']['ACCLASS'])
    acc_loc.append(data['features'][i]['attributes']['ACCLOC'])
    acc_district.append(data['features'][i]['attributes']['District'])
    acc_division_code.append(data['features'][i]['attributes']['Division'])
    acc_date.append(data['features'][i]['attributes']['DATE'])
    acc_invtype.append(data['features'][i]['attributes']['INVTYPE'])
    acc_invage.append(data['features'][i]['attributes']['INVAGE'])
    acc_injur.append(data['features'][i]['attributes']['INJURY'])
    acc_light.append(data['features'][i]['attributes']['LIGHT'])
    acc_impactype.append(data['features'][i]['attributes']['IMPACTYPE'])
    acc_longitue.append(data['features'][i]['attributes']['LONGITUDE'])
    acc_latitute.append(data['features'][i]['attributes']['LATITUDE'])
    acc_road_condition.append(data['features'][i]['attributes']['RDSFCOND'])
    acc_time.append(data['features'][i]['attributes']['TIME'])
    acc_traffic.append(data['features'][i]['attributes']['TRAFFCTL'])
    acc_visibility.append(data['features'][i]['attributes']['VISIBILITY'])
    acc_year.append(data['features'][i]['attributes']['YEAR'])
```

```python
data_dict = {
    'Class': acc_class,
    'Location': acc_loc,
    'District': acc_district,
    'Division Code': acc_division_code,
    'Longitue': acc_longitue,
    'Latitute': acc_latitute,
    'Data': acc_date,
    'Year': acc_year,
    'Injure': acc_injur,
    'Injure Type': acc_invtype,
    'Injure Age': acc_invage,
    'Light': acc_light,
    'Imactype': acc_impactype,
    'Road Condition': acc_road_condition,
    'Traffic': acc_traffic,
    'Visibility': acc_visibility,
}
```

```python
df = pd.DataFrame(data_dict)
df.head()
```
