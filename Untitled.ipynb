```python
import os
import csv
import pandas as pd
```

```python
#loading and reading the csv file
crimes_csv = "../toronto-crime-analysis/crimes_csv.csv"
crimes_pd = pd.read_csv(crimes_csv)
crimes_pd.head()
```

```python
#calculating the most common crimes in Toronto
crime_type = []
crime_count = []

most_common_crimes_count = crimes_pd.groupby('MCI').size()
for i in range(len(most_common_crimes_count)):
    crime_type.append(most_common_crimes_count[0])
    crime_count.append(most_common_crimes_count[1])
most_common_crimes = pd.DataFrame({'Crime Type' : crime_type,
                                   'Crime Count' : crime_count
                                  })
most_common_crimes
```

```python
#
highest_crime_rates = crimes_pd.groupby('Neighbourhood').size()
highest_crime_rates = highest_crime_rates.sort_values(ascending=False)
print(highest_crime_rates)
```
