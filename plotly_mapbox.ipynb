```python
data_sample = pd.read_csv('data_sample.csv')
data_sample.head()
```

```python
map_data = data_sample[['Hood ID', 'Latitute', 'Longitue', 'District']]
len(map_data)
```

```python
test = map_data.drop_duplicates()
test.sort_values('Hood ID')
```

```python
map_data_sort = map_data.sort_values('Hood ID')
map_dict = {
    'id': map_data['Hood ID'].unique(),
    'lan': map_data.groupby('Hood ID')['Latitute'].mean(),
    'lon': map_data.groupby('Hood ID')['Longitue'].mean(),
    'frequency': map_data.groupby('Hood ID')['Hood ID'].count(),
    'district': map_data.groupby('Hood ID')['District'].unique()
}
```

```python
map_dict_df = pd.DataFrame(map_dict)
map_dict_df.head()
```

```python
map_plot_list = []
for index, row in map_dict_df.iterrows():
    
    data = [
        go.Scattermapbox(
            lat=[str(row['lan'])],
            lon=[str(row['lon'])],
            mode='markers',
            marker=dict(
                size=row['frequency']
            ),
            text=row['district'],
        )
    ]
    
    map_plot_list.append(data[0])
    

pprint(map_plot_list[0: 3])
```

```python
fig = dict(data=map_plot_list, layout=layout)

py.iplot(fig, filename='Toronto Crime')
```
